```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
numbers = []
with open('bad_si_sdr.txt', 'r') as f:
    for line in f.readlines():
        numbers.append(float(line))
```

```python
weights = np.ones_like(numbers) / len(numbers)
plt.hist(numbers, weights=weights, bins=20)
plt.xlabel('si_sdr')
plt.ylabel('density')
percent_of_negatives = len([x for x in numbers if x < 0]) / len(numbers)
print(f'mean={np.mean(numbers)}', f'median={np.median(numbers)}', f'percent_of_negatives={percent_of_negatives}', sep='\n')
```
